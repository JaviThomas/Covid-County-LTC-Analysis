```python
f = './Reporting_data/IL_2020-11-13_Facilities_LTC_data_v2.csv'
IL_state_data = pd.read_csv(f,  parse_dates=['ReportingDate'], infer_datetime_format=True) 
IL_state_data['FacilityName'] = IL_state_data['FacilityName'].apply(lambda x: x.strip().replace('  ', ' ').upper())
IL_state_data.rename(columns={'FacilityName': 'Provider Name'}, inplace=True)
IL_state_data.sort_values(by=['County', 'Provider Name'], inplace=True)
```

```python
cols=['County', 'Provider Name', 'VLOOKUP', 'Change Score']
IL_state_data['VLOOKUP'] = '' # For Spreadsheet
IL_state_data['Change Score'] = '' # For Spreadsheet
for county in list(IL_cms_data['County'].unique()):
    #print(county)
    # Get lists of Facilities for the counties
    df_state = IL_state_data[IL_state_data['County'] == county]
    df_cms = IL_cms_data[IL_cms_data['County'] == county]
    state_facs = list(df_state['Provider Name'])
    cms_facs = list(df_cms['Provider Name'])
    for str2Match in state_facs:
        try:
            Ratios = process.extract(str2Match,cms_facs, limit=2)
            IL_state_data.loc[(IL_state_data['County'] == county) & (IL_state_data['Provider Name'] == str2Match), 'VLOOKUP'] = Ratios[0][0]            
            IL_state_data.loc[(IL_state_data['County'] == county) & (IL_state_data['Provider Name'] == str2Match), 'Change Score'] = Ratios[0][1]
        except:
            pass

IL_state_data['Change Score'].value_counts()
# str2Match = cms_facs[0]
# strOptions = state_facs

# # Pass 1 Remove Exact Matches
# for str2Match in cms_facs:
#     Ratios = process.extract(str2Match,strOptions, limit=3)
#     print(str2Match, Ratios)
```
