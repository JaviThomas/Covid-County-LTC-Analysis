```python
import pandas as pd
import json
import glob
import IPython.display
import altair as alt

pd.options.display.max_columns = None
```

```python
for file in glob.glob('Source_data/IL_*LTC_data_Source.json'):
    print(file)
```

```python

with open('Source_data/IL_2021-01-15_LTC_data_Source.json') as fp:
    data = fp.read()
    jsonData = json.loads(data)
```

```python
df = pd.DataFrame(jsonData['LTC_Historical_Reported_Cases'])
df.sort_values(by=['reported_date'], inplace=True)
df['new_deaths']= df.deaths.diff()
df['new_confirmed_cases'] = df.confirmed_cases.diff()
```

```python
chart_title = ' Cases by Week'
chart1 = alt.Chart(df).mark_bar().encode(
                     x='reported_date:O',
                     y='new_confirmed_cases',
                     ).properties( width=400, height=400, title=chart_title)
chart1
```
