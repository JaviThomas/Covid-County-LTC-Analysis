```python
import pandas as pd
import urllib3 as urllib
import json

import IPython.display
import altair as alt

pd.options.display.max_columns = None
```

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRa9HnmEl83YXHfbgSPpt0fJe4SyuYLc0GuBAglF4yMYaoKSPRCyXASaWXMrTu1WEYp1oeJZIYHpj7t/pub?gid=467018747&single=true&output=csv'
df = pd.read_csv(url, skiprows=1, parse_dates=['Date'])
df.sort_values(by=['State', 'Date'], inplace=True)
```

```python
df.columns
```

```python
cols = ['Date', 'State', 'Data Set', 'Resident Positives', 'Resident Deaths',
       'Staff Positives', 
       'Resident/Staff Deaths']

# Weekly Resident Confirmed COVID-19 Cases Per 1,000 Residents
# Weekly Resident COVID-19 Deaths Per 1,000 Residents
# Total Resident Confirmed COVID-19 Cases Per 1,000 Residents
states = list(sorted(df['State'].unique()))
field = cols[4]
for state in states:
    chart_title = state + ': ' + field
#     state_df = df[(df['Date'] != min(df['Date'])) & (df['State'] == state)].groupby('Date').sum()
    state_df = df[(df['State'] == state)].groupby('Date').sum()
    state_df['new_'+field]= state_df[field].diff()
    state_df['Date'] = state_df.index

    chart1 = alt.Chart(state_df).mark_bar().encode(
                         x=alt.X('yearmonthdate(Date):O', title='date', axis=alt.Axis(labelSeparation=10)),
                         y='new_'+field,
                         ).properties( width=200, height=200, title=chart_title)
    display(chart1)
```
