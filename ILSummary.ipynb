```python
import glob
import json
import pandas as pd

from IPython.display import JSON
```

```python
#glob.glob('Reporting_data/IL_*_Facilities_LTC_data_v2.csv'):
df = pd.concat([pd.read_csv(f,  parse_dates=['ReportingDate'], infer_datetime_format=True) for f in glob.glob('Reporting_data/IL_*_Facilities_LTC_data_v2.csv')], ignore_index = True)
```

```python
df.sort_values(by=['County', 'FacilityName', 'ReportingDate'], inplace=True)
```

```python
df['NewConfirmedCases'] = df.groupby(['County', 'FacilityName'])['confirmed_cases'].apply(lambda x: x - x.shift(1))
df['NewDeaths'] = df.groupby(['County', 'FacilityName'])['deaths'].apply(lambda x: x - x.shift(1))
df['NewOutbreaks'] = df.groupby(['County', 'FacilityName'])['outbreaks'].apply(lambda x: x - x.shift(1))
df['DaysDiff'] = df.groupby(['County', 'FacilityName'])['ReportingDate'].apply(lambda x: x - x.shift(1))

reporting_date = str(df.ReportingDate.max())[0:10]
filename =  "Reporting_data/IL_" + reporting_date + "FullSummary_LTC_data_Source.csv"
df.to_csv(filename, index=False)
filename
```
