```python
import numpy as np
import pandas as pd
import urllib.request as urllib2
import json
import IPython.display
```

# 1 - Get CMS Data and load into DataFrame

```python
url_csv = 'https://data.cms.gov/api/views/s2uc-8wxp/rows.csv?accessType=DOWNLOAD&api_foundry=true'

response = urllib2.urlopen(url_csv)
#cms_data = pd.read_csv(response, parse_dates=['Week Ending'], dtype={'Provider Name': str, 'Federal Provider Number': str})
cms_data = pd.read_csv(response, dtype={'Provider Name': str, 'Federal Provider Number': str})
max_date = cms_data['Week Ending'].max()
reporting_date = str(max_date)[0:10]
print(reporting_date)
display(cms_data.head(5))
```

# 2 - Create DataFrame of CMS data for a particular state (IL)

```python
my_state = 'IL'
IL_cms_data = cms_data[(cms_data['Provider State'] == my_state)]
IL_cms_data_latest = cms_data[(cms_data['Week Ending'] == max_date) & (cms_data['Provider State'] == my_state)]
IL_cms_data['Facility-County'] = IL_cms_data['Provider Name'] + '-' + IL_cms_data['County']
```
